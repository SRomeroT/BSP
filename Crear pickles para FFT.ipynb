```python
import math
import numpy as np
import matplotlib.pyplot as plt
import pickle
import pandas as pd
from numpy.fft import fft, fftfreq
```

```python
part = "C"

signal_name_vd = "vdTrozo" + part
signal_name_lp = "lpTrozo" + part

vd_directory = signal_name_vd + ".pickle"
lp_directory = signal_name_lp  + ".pickle"
```

```python
vd = pickle.load(open(vd_directory, "rb" ))
lp = pickle.load(open(lp_directory , "rb" ))

vd = np.array(np.char.replace(vd, ',', '.'),dtype='double')
lp = np.array(np.char.replace(lp, ',', '.'),dtype='double')
```

```python
partR = "R"

signal_nameR_vd = "vdTrozo" + partR
signal_nameR_lp = "lpTrozo" + partR

vd_directoryR = signal_nameR_vd + ".pickle"
lp_directoryR = signal_nameR_lp  + ".pickle"
```

```python
vd_R = pickle.load(open(vd_directoryR, "rb" ))
lp_R = pickle.load(open(lp_directoryR , "rb" ))

vd_R = np.array(np.char.replace(vd_R, ',', '.'),dtype='double')
lp_R = np.array(np.char.replace(lp_R, ',', '.'),dtype='double')
```

```python
partG = "G"

signal_nameG_vd = "vdTrozo" + partG
signal_nameG_lp = "lpTrozo" + partG
signal_nameG_tyd = "tydTrozo" + partG

vd_directoryG = signal_nameG_vd + ".pickle"
lp_directoryG = signal_nameG_lp  + ".pickle"
tyd_directoryG = signal_nameG_tyd  + ".pickle"
```

```python
vd_G = pickle.load(open(vd_directoryG, "rb" ))
lp_G = pickle.load(open(lp_directoryG , "rb" ))
tyd_G = pickle.load(open(tyd_directoryG , "rb" ))

vd_G = np.array(np.char.replace(vd_G, ',', '.'),dtype='double')
lp_G = np.array(np.char.replace(lp_G, ',', '.'),dtype='double')
tyd_G = np.array(np.char.replace(tyd_G, ',', '.'),dtype='double')
```

```python
plt.plot(tyd_G[:1000000])
plt.show()
```

```python
pd_control = pd.DataFrame({'vd': vd, 'lp': lp})
pd_control
```

```python
pd_recuperacion = pd.DataFrame({'vd': vd_R, 'lp': lp_R})
pd_recuperacion
```

```python
pd_gaba = pd.DataFrame({'vd': vd_G, 'lp': lp_G, 'tyd':tyd_G})
pd_gaba
```

```python
pd_gaba
pd_recuperacion
pd_control

with open('pd_gaba.pickle', 'wb') as handle:
    pickle.dump(pd_gaba, handle, protocol=pickle.HIGHEST_PROTOCOL)
    
with open('pd_recuperacion.pickle', 'wb') as handle:
    pickle.dump(pd_recuperacion, handle, protocol=pickle.HIGHEST_PROTOCOL)
    
with open('pd_control.pickle', 'wb') as handle:
    pickle.dump(pd_control, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
